The goal of [COVID-19 Data Hub](https://covid19datahub.io/) is to provide the research community with a unified data hub by collecting worldwide fine-grained case data, merged with exogenous variables helpful for a better understanding of COVID-19.

Python package `covid19dh` brings the simple access to COVID-19 Data Hub. Acquiring the data is as simple as

```python
from covid19dh import covid19
usa1 = covid19("USA") # read US data
```

```python
usa2 = covid19("USA", level = 2, verbose = False) # fetch states data
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('figure', figsize=(12, 5))
```

```python
# plot confirmed
plt.plot(usa1["date"], usa1["confirmed"], "r")
# format plot
plt.gcf().autofmt_xdate()
plt.suptitle("Confirmed cases of Covid-19 in USA", size=16)
plt.show()
```

```python
# plot confirmed
plt.plot(usa1["date"], usa1["deaths"], "k")
# format plot
plt.gcf().autofmt_xdate()
plt.suptitle("Deaths for Covid-19 in USA", size=16)
plt.show()
```

```python
# plot confirmed
plt.plot(usa1["date"], usa1["recovered"], "y")
# format plot
plt.gcf().autofmt_xdate()
plt.suptitle("Recovered of Covid-19 in USA", size=16)
plt.show()
```

```python
# plot each state separately
for state in usa2.groupby("administrative_area_level_2").groups:
    data = usa2[usa2["administrative_area_level_2"] == state]
    plt.plot(data["date"], data["confirmed"], label = state)
# format plot
plt.gcf().autofmt_xdate()
plt.suptitle("Confirmed cases of Covid-19 in US States", size = 16)
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.25), fancybox = False, ncol = 8, prop={'size': 9})
plt.show()
```
